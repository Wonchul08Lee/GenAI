```python
import numpy as np
from transformers.trainer_utils import EvalPrediction

def accuracy_score(predict):
    preds = np.argmax(predict.predictions, axis=1)
   
    acc = (preds == predict.label_ids).mean()
    return {"accuracy": acc}

y_hat = np.array([[0.2, 0.8], [0.4, 0.6], [0.1, 0.9]])
y_true = np.array([1, 0, 1])

pred = EvalPrediction(predictions=y_hat, label_ids=y_true)
print(pred.predictions)
print(pred.label_ids)

print(vars(pred))

print(pred.__dict__)
accuracy_score(pred)
```

```python
import  torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification

#model_name  = "beomi/kcbert-base"
model_name = "monologg/koelectra-base-v3-finetuned-korquad"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2)
```

```python
text = "이 영화 너무 좋았어! 최고야!"
inputs = tokenizer(text, return_tensors="pt")
```

```python
model.eval()
with torch.no_grad():
    outputs = model(**inputs)
```

```python
logits = outputs.logits
sentiment = torch.argmax(logits, dim=1).item()
print("결과:", "긍정" if sentiment == 1 else "부정")
```
